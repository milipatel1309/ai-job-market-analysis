```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../data/raw/ai_job_dataset.csv")
df.head()
```

```python
df.shape
```

```python
df.columns
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.columns = df.columns.str.lower().str.strip().str.replace(" ", "_")
df.columns
```

```python
df = df.drop_duplicates()
df.shape
```

```python
df.isnull().sum()
```

```python
df["required_skills"] = df["required_skills"].astype(str)
df["skills_list"] = df["required_skills"].apply(lambda x: [skill.strip() for skill in x.split(",")])
df[["required_skills", "skills_list"]].head()
```

```python
# Final check before saving cleaned data
print("Final shape:", df.shape)
print("Missing values:", df.isnull().sum().sum())

# Check duplicates only on original/simple columns, not skills_list
print("Duplicate rows:", df.drop(columns=["skills_list"]).duplicated().sum())

df.head()
```

```python
df.to_csv("../data/processed/ai_job_dataset_cleaned.csv", index=False)
```
