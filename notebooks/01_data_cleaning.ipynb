```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../data/raw/ai_job_dataset.csv")
df.head()
```

```python
df.columns = df.columns.str.lower().str.strip().str.replace(" ", "_")
df.columns
```

```python
df = df.drop_duplicates()
df.shape
```

```python
df.isnull().sum()
```

```python
df["required_skills"] = df["required_skills"].astype(str)
df["skills_list"] = df["required_skills"].apply(lambda x: [skill.strip() for skill in x.split(",")])
df[["required_skills", "skills_list"]].head()
```

```python
df.to_csv("../data/processed/ai_job_dataset_cleaned.csv", index=False)
```
